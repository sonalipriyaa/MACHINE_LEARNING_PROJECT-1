IMPORTING THE DEPENDENCIES

```python
import numpy as np; #numpy array will be used to make numpy array its like list in python
import pandas as pd; # used to create dataframes they are like structured tables
from sklearn.model_selection import train_test_split; # original data will be split into testing and training data
from sklearn.linear_model import LogisticRegression;
from sklearn.metrics import accuracy_score; # will be used to check how our model will perform
```

DATA COLLECTION AND PROCESSING

```python
# will now load csv datafile into pandas dataframe
heart_data=pd.read_csv('/content/heart.csv');
```

```python
# head function will print first 5 rows of pandas dataframe
heart_data.head()
```

```python
# printing last 5 rows
heart_data.tail()
```

```python
# NO OF ROWS AND COL IN DATASET
heart_data.shape
# 1025 rows and 14 col
```

```python
#info about the data
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# statistical measures of the data like count standard deviation mean etc
heart_data.describe()
```

```python
#checking target value count as 0 or 1 total value
# here distribution is almost equal
heart_data['target'].value_counts()
# 0--> no heart problem
# 1--> heart problem
```

SPLITING THE FEATUES/ATRIBUTES AND TARGET

```python
# using all featues to predict the target
# storing features in variable X and target in variable Y for training and testing our model
# when droping a column axis will be 1 and for row it will be 0 so removing target from featues
X = heart_data.drop(columns='target' , axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

SPLITING DATA INTO TRAIN DATA AND TEST DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y,random_state=2)
#20 % data used as test data here
# two classes that is 0 and 1 will be distributed in even manner by the use of stratify
```

```python
#checking number of training and test data
print(X.shape,X_test.shape,X_train.shape)
```

TRAINING OUR MACHINE LEARNING MODEL USING LOGISTIC REGRESSION

```python
# in variable named model we will load our logistic regression model
model = LogisticRegression()
```

TRAINING LOGISTIC REGRESSION MODEL WITH TRAINING DATA

```python
model.fit(X_train,Y_train)
#will match features and corresponding targets will find relationship between features and target
```

MODEL EVALUATION USING ACCURACY SCORE

```python
# model will be asked to predict the value and this predicted value will be compared with the original value
# first finding accuracy on the training data
X_train_prediction=model.predict(X_train)
# we have trained our model with x_train and y_train only so will give x_train and will ask to predict the target value that is y_train and i am storing that target value in x_train prediction latter will be comparing it with original value
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy on traing data is: ',training_data_accuracy)
```

```python
# accuracy on test data
# now will find target for x_test and compare predicted value from y_test
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('accuracy on test data is: ',test_data_accuracy)
```

BUILDING A PREDICTIVE SYSTEM


```python
# take input value as features
# change the input data into a numpy array
input_data=(58,0,0,100,248,0,0,122,0,1,1,0,2)
input_data_as_numpy_array=np.asarray(input_data);
# reshaping to tell that we are predicting it for only one value
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)
prediction=model.predict(input_data_reshape)
print(prediction)
if(prediction[0] == 0):
    print('person is healthy')
else:
  print('the person has heart disease')
```
